```python
import json
import pandas as pd
import glob

json_files = glob.glob("/Users/kuyesura/workplace/Code/DE_Prep/Airflow/fda_drug_shortages/data/**/*.json", recursive=True)

dfs = []
for f in json_files:
    try:
        with open(f, 'r') as file:
            data = json.load(file)
            df_temp = pd.json_normalize(data)  # Flattens nested JSON
            dfs.append(df_temp)
    except Exception as e:
        print(f"Error reading {f}: {e}")

if dfs:
    df = pd.concat(dfs, ignore_index=True)

```
